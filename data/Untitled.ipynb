```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# CSV location 

summer = 'SummerOlympicMedals.csv'
```

```python
#Reading CSV, Selecting columns, Renaming 
df_summer = pd.read_csv(summer, encoding="ISO-8859-1")
df_summer = df_summer[["City","Edition","Sport","Discipline","Athlete","NOC","Gender","Event","Medal"]]
df_summer = df_summer.rename(columns={"City":"city_host", "Edition":"year","Sport":"sport","Discipline":"discipline","Athlete":"athlete","NOC":"country_code","Gender":"gender","Event":"event","Medal":"medal"})
df_summer.head()
```

```python
#Identifying unique discipline
discipline_list = []
for i in range(0, len(df_summer)):
    if isinstance(df_summer['discipline'][i] str):
        discipline = df_summer['discipline'][i].split(', ')
        for j in range(0, len(discipline)):
            if discipline[j] not in discipline_list:
                discipline_list.append(discipline[j])

#Creating columns in df with all unique genres
for discipline in discipline_list:
    discipline['discipline_'+ discipline] = np.nan
```
